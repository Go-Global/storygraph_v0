```python
import pandas as pd

df = pd.read_csv('../data/realdonaldtrump.csv')
df.head()
```

```python
from neo4j import GraphDatabase
from dotenv import dotenv_values
from soup_models import Document, Source
from collections import defaultdict

config = dotenv_values(".env")
```

```python
from graph_driver import GraphDBDriver
driver = GraphDBDriver(uri=config["REMOTE_GRAPH_URI"], user=config["REMOTE_GRAPH_USER"], password=config["REMOTE_GRAPH_PWD"])
```

```python
# Source(key, name, source_type)
trump = Source('25073877', 'realdonaldtrump', 'twitter-user')
str(trump)
```

```python
import datetime

tweets = []
for index, row in df.iterrows():
    #Document(self, key, title, doc_type, attrs=dict(), spacy_doc=None, date_processed=None):
    tweet_date = datetime.datetime.strptime(row.date, "%Y-%m-%d %H:%M:%S")
#     tweet_date = datetime.datetime.fromisoformat(row.date)
    attrs = {'created_at': tweet_date, 'retweets':row.retweets, 'favorites':row.favorites, 'mentions':row.mentions, 'hashtags':row.hashtags, 'link': row.link}
    tweet = Document(str(row.id), row.content, 'tweet', attrs=attrs, date_processed=datetime.datetime.now())
    tweets.append(tweet)

# row = df[0]
# row
```

```python
driver.upload_nodes(tweets[16000:])
```

```python
from soup_models import Authored
edges = [Authored(trump, tweet) for tweet in tweets]
```

```python
# for edge in edges:
#     edge.attrs = dict(edge.attrs)
edge = Authored(trump, tweets[0])
edge.attrs = dict(edge.attrs)
```

```python
def upload_edge(edge):
    source_key = edge.source_key
    dest_key = edge.dest_key
    print(source_key, dest_key)
    query = ["MATCH (a), (b)", "WHERE a.key=\"{}\" AND b.key = \"{}\"".format(source_key, dest_key), "CREATE (a)-[edge:{}]->(b)".format('authored')]
    return driver.query(" ".join(query))
print(upload_edge(edge))
```

```python
driver.upload_edges(edges)
```
